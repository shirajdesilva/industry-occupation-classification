```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('OSCA Category Descriptions.xlsx', sheet_name='Table 1', skiprows=4)    
df.head()
```

```python
df.shape
```

```python
# Concatenate Principal Title and Alternative Title
df['OCSA TItle'] = df['Principal Title'].astype(str) + ' , ' + df['Alternative Title'].astype(str)
```

```python
# Remove all columns except 'OCSA Code' and 'OCSA Title'
df = df[['Identifier', 'OCSA TItle', 'Principal Title','Alternative Title', 'Lead Statement','Main Tasks','Specialisations']]
df.head()
```

```python
# export to parquet
df['Identifier'] = df['Identifier'].astype(str)
df.to_parquet('OSCA_Category_Descriptions.parquet', index=False)
```
